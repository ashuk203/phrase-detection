```python
num_matches = 0
matches = []

for prdr in prdualrank_words:
    prdr_tok = nlp(prdr.lower().replace("-", " "))
    for ap in autophrase_words:
        ap_tok = nlp(ap.lower().replace("-", " "))
        similarity = ap_tok.similarity(prdr_tok) * 100
        if similarity >= 80.0:
            num_matches += 1
            matches.append((prdr, ap))
            break
```
